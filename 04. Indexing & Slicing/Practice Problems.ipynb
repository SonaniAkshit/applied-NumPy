***1. Select the 3rd element from a 1D array***

```python
arr = np.arange(21,80)
print(arr)
print(f"3rd element:{arr[2]}")
```

***2. Get last element using negative indexing***


```python
arr = np.arange(41,71)
print(arr[-1])
```

***3. Select first row from 2D array***

```python
mat = np.arange(1,26)
mat = mat.reshape(5,5)
print(f"Matrix:{mat}\n")
print(f"First Row: {mat[0:1]}")
```


***4. Select second column from matrix***

```python
# use same matrix

print(mat[:,1:2])
```


***5. Slice first 5 elements from array***

```python
arr = np.arange(1,21)
print(f"Full Array: {arr}")
print(f"Fisrt 5 Elements: {arr[:5]}")
```
