### **Problem 1: Epoch Counter**

You train a model for **50 epochs** and want an array that stores the **epoch numbers starting from 0**.

**Task:**
Create the array using `np.arange()`.


```python
x = np.arange(0,50)
x
```

### **Problem 2: Sample Indexing**

You have **120 data samples** and need an array of **sample indices starting from 1**.

**Task:**
Create the array using `np.arange()`.


```python
x = np.arange(1,121)
x
```

### **Problem 3: Time Steps**

A simulation runs for **10 seconds** with a **time step of 0.5 seconds**, starting from **0**.

**Task:**
Create a NumPy array of time steps using `np.arange()`.


```python
x = np.arange(0,10,0.5)
x
```

### **Problem 4: Feature Column IDs**

A dataset has **15 features**, and you want an array of **feature IDs from 0 to 14**.

**Task:**
Create the array using `np.arange()`.


```python
x = np.arange(0,15)
x
```

### **Problem 5: Even Number Generator**

You need an array of **even numbers from 2 up to (but not including) 40**.

**Task:**
Create the array using `np.arange()`.


```python
x = np.arange(2,40,2)
x
```
