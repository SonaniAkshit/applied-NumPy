# 6. np.linspace()

Creates evenly spaced values between two limits.

```python
import numpy as np
```

```python
x = np.linspace(1,5,10)
x
```

```python
x = np.linspace(0.01,0.1,10)
x
```


### **Problem 1: Learning Rate Schedule**
