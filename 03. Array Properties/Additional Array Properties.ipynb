# Additional Array Properties

- `itemsize` 
- `nbytes` 
- `T`  (transpose)
- `strides` 
- `flags`

> These properties describe **memory layout and data movement**, not just shape.

### ***Syntax & Examples***
We use two different array and inspect all properties

```python
import numpy as np
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]], dtype=np.int32)
```

```python
x = np.array([1.5, 2.5])
```

# 1. `itemsize`

```python
arr.itemsize
```

Explanation:
- int32 = 4 bytes
- Each element uses 4 bytes

```python
x.itemsize
```

Explanation:
- float64
- Higher precision, more memory

# 2. `nbytes`

```python
arr.nbytes
```

Explanation:
- 6 elements
- 6 × 4 bytes = 24

```python
x.nbytes
```

Explanation:
- 2 × 8 bytes

# 3. `T`(Transpose)

```python
# original arrays 

print(arr)
print(x)
```

```python
# Transpose

print(arr.T)
print(arr.shape)
print(arr.T.shape)
```

Explanation:
- Shape changed from (2,3) to (3,2)
- Data not copied

```python
print(x.T)
```

# 4. `strides` 

```python
print(arr.strides)
```

Explanation:
- Move 12 bytes to next row
- Move 4 bytes to next column
 

# 5.`flags`

```python
print(arr.flags)
```

Explanation:

- Stored row-wise
- Can be modified
- Owns its memory
