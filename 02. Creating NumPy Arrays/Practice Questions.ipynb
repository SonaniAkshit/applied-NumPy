**9. Why is np.arange(0,1,0.1) unsafe?**

Because of floating-point precision.
`0.1` cannot be represented exactly in binary, so the result can miss or add values unpredictably.
Safer choice: `np.linspace(0, 1, 10)`.


**10. When should np.empty() be used?**

When you want **fast allocation** and you will **overwrite every element** immediately.
It does not initialize values, so reading before writing gives garbage data.


**11. Difference between (5,) and (5,1)?**

- `(5,)` → 1D array with 5 elements

- `(5,1)` → 2D array with 5 rows and 1 column

This matters for broadcasting, matrix multiplication, and ML models.


**12. What happens if mixed datatypes are passed?**

NumPy upcasts to a common type.

Example: `int + float → float`, `number + string → string`.

This can silently increase memory usage or break numeric operations.


**13. Why is np.zeros common in ML?**

Because many ML algorithms **start parameters, weights, gradients, or buffers at zero**.
It gives a known, clean baseline and predictable behavior.

### `Hard (15–20)`

**15. Why can bad initialization break gradient descent?**

It can cause **vanishing gradients**, **exploding gradients**, or **symmetry problems**.
Result: weights do not learn, training stalls, or loss becomes unstable.

---

**16. Memory difference between list and NumPy array?**

Python lists store **references to objects**.
NumPy arrays store **contiguous raw values** of a single type.
Result: NumPy uses less memory and is much faster for math.

---

**17. Why identity matrices matter in linear models?**

They **preserve input values** while controlling transformations.
Used in regularization, initialization, and to avoid changing feature scale unintentionally.

---

**18. How wrong shapes silently break ML models?**

Broadcasting may apply operations **without raising errors**, but on the wrong dimensions.
Model trains, but learns incorrect patterns.

---

**19. When does `np.empty()` cause production bugs?**

When code **assumes zero-initialization**.
Uninitialized memory leads to random values, nondeterministic behavior, and hard-to-reproduce bugs.

---

**20. `arange` vs `linspace` in numerical stability?**

* `arange`: step-based, prone to floating-point errors
* `linspace`: count-based, numerically stable
  Use `linspace` when precision matters.
